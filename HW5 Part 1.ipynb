```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
x1=0
x2=np.sqrt(0.75)*np.random.randn(1,1)+x1/2+0.5
for i in range(5000):
    x1=np.append(x1,np.sqrt(0.75)*np.random.randn()+x2[i]/2+0.5)
    x2=np.append(x2,np.sqrt(0.75)*np.random.randn()+x1[i+1]/2+0.5)
```

```python
%matplotlib inline
cont,bins,ignored=plt.hist(x1,40,normed=True)
plt.plot(bins,1/np.sqrt(2*np.pi)*np.exp(-(bins-1)**2/2),linewidth=2,color='r')
```

```python
cont,bins,ignored=plt.hist(x2,40,normed=True)
plt.plot(bins,1/np.sqrt(2*np.pi)*np.exp(-(bins-1)**2/2),linewidth=2,color='r')
```
